```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_classification
from sklearn.datasets import make_blobs
```

```python
data, label  = make_blobs(n_samples=[20,100],
                                   n_features=2,
                                   centers=[[-1,-1],[1,1]],
                                   cluster_std=1)
color = ['blue','red']

plt.scatter(data[:,0],data[:,1],c=label)
```

```python
from imblearn.combine import SMOTETomek
smt = SMOTETomek()
data_res, label_res = smt.fit_resample(data,label)
plt.scatter(data_res[:,0],data_res[:,1],c=label_res)
```
