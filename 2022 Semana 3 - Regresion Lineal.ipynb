```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('/content/Regresion Lineal.csv')
```

```python
data
```

```python
x = data.X.to_numpy().reshape(-1, 1)
y = data.Y.to_numpy().reshape(-1, 1)
```

```python
plt.figure()
plt.plot(x,y,'b.')
plt.show()
```

```python
from sklearn.linear_model import LinearRegression
```

```python
model = LinearRegression()
model.fit(x,y)
model.score(x,y)
```

```python
print(model.coef_)
print(model.intercept_)
```

```python
x_predict = np.array([-1]).reshape(-1, 1)

model.predict(x_predict)
```

```python
casas = pd.read_csv('https://raw.githubusercontent.com/ageron/handson-ml/master/datasets/housing/housing.csv')
casas.drop(labels=['ocean_proximity'],axis=1,inplace=True)
casas.dropna(inplace=True)
casas
```

```python
x_casas = casas.drop(labels=['median_house_value'],axis=1).to_numpy()
y_casas = casas.median_house_value.to_numpy().reshape(-1, 1)
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(x_casas,y_casas,test_size=0.3,random_state=420)
```

```python
print(X_train.shape)
print(X_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
model_casas = LinearRegression()
model_casas.fit(X_train,y_train)
model_casas.score(X_test,y_test)
```

```python
print(model_casas.coef_)
print(model_casas.intercept_)
```
